#### ***Why do we need to set the PREFECT_HOME environment variable?
We set the PREFECT_HOME environment variable so Prefect knows where to store all of its local files.
This directory contains:
- the local SQLite database
- configuration files
- logs
- cached blocks
- storage for flows
- internal runtime metadata

This way Prefect uses a clean, project-specific location for its database and metadata instead of the default global one.

## **2.5 Performing a prediction**
***What's the purpose of the MLFlow Model Registry? Why do you need to use the registered model from MLFlow and not the model file directly?***
- Purpose of the MLFlow Model Registry:

The MLFlow Model Registry is like a centralized hub for all your machine learning models.
- Versioning
  - Every time you register a new model (or a new version of an existing model), MLFlow keeps track of it.

- Centralized access
  - Instead of using a local model file, teams can load the same model from the registry from anywhere.

- Lifecycle management
  - Models can have stages like Staging, Production, or Archived, which helps in deployment pipelines.

- Consistency and reproducibility
  - Ensures that everyone (or any pipeline) uses the exact same trained model, avoiding works on my machine problems.

- Why use the registered model instead of the local file?
  - The registered model guarantees that you are using a production-ready, versioned model, not just a local file that may be outdated.
  - It integrates directly with MLFlows tracking and deployment features (loading it in other environments, serving via MLFlow Models API, etc.).
